```python
import tarfile
from io import StringIO
import pandas as pd
tar = tarfile.open("adult.tar.gz", "r:gz")
for member in tar.getmembers():
    f = tar.extractfile(member)
    if(f is not None):
         content = f.read()
         

s=str(content,'utf-8')
data = StringIO(s) 
df=pd.read_csv(data)
df
```
